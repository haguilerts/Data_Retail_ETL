```python



# 1.1 Cargar los archivos CSV en DataFrames distintos
customer_df = pd.read_csv('customer_data.csv')
sales_df = pd.read_csv('sales_data.csv')

# 1.2 Mostrar las primeras filas para verificar la carga
print("Datos de clientes:")
#print(customer_df.head(10))
print("\nDatos de ventas:")
#print(sales_df.head(10))

# 1.3 Describir el proceso de extracción
print("\nInformación de los DataFrames:")
#print(customer_df.info())
#print(sales_df.info())

# 1.4 Concatenar ambos DataFrames (usando merge por el campo común 'customer_id')
merged_df = pd.merge(sales_df, customer_df, on='customer_id', how='inner')
print("\nDataFrame combinado (ventas + clientes):")
print(merged_df.head(20))

```

```python
# 1. Crear una máscara booleana para identificar a TODOS los customer_id duplicados
#    (keep=False marca la primera, la última y todas las ocurrencias repetidas)
mascara_clientes_repetidos = merged_df.duplicated(subset=['customer_id'], keep=False)

# 2. Usar esa máscara para filtrar el DataFrame y obtener solo las filas deseadas
clientes_con_mas_de_una_compra = merged_df[mascara_clientes_repetidos]

# 3. Mostrar el resultado
print("Filas completas de los clientes que han comprado más de una vez:")
print(clientes_con_mas_de_una_compra)
# 1. Contar cuántas veces aparece cada customer_id
conteo_ids = merged_df['customer_id'].value_counts()

# 2. Filtrar para ver solo aquellos IDs con un conteo mayor a 1
ids_repetidos_confirmados = conteo_ids[conteo_ids > 1]

print("Conteo de IDs de clientes que compraron más de una vez:")
print(ids_repetidos_confirmados)

# 1. Contar cuántas veces aparece cada invoice_no
conteo_invoices = merged_df['invoice_no'].value_counts()

# 2. Filtrar para ver solo aquellos invoice_no con un conteo mayor a 1 (es decir, que están repetidos)
invoices_repetidas_confirmadas = conteo_invoices[conteo_invoices > 1]

print("Conteo de números de factura (invoice_no) que se repiten:")
print(invoices_repetidas_confirmadas)
```

```python
# 1. CONSULTA DE NULOS (Buena práctica para saber cuántos se eliminarán)
print("\nValores nulos por columna ANTES de limpiar:")
print(merged_df.isnull().sum()) # Nos dice cuántos nulos hay en 'age'

# 2. CONTEO INICIAL
# Usa merged_df.shape[0] para el número total de filas, incluyendo nulos
print("INICIO--- cantidad total de filas: ", merged_df.shape[0]) 

# 3. LIMPIEZA
merged_df = merged_df.dropna()

# 4. CONTEO FINAL
# Cuenta el número de filas en el DataFrame limpio
print("FIN----cantidad total de filas: ", merged_df.shape[0]) 

# 5. VERIFICACIÓN FINAL
print("\nValores nulos por columna DESPUÉS de limpiar:")
print(merged_df.isnull().sum())
```

```python
# 2.1 Limpieza de valores nulos
print("\nValores nulos por columna:")
print(merged_df.isnull().sum())
merged_df = merged_df.dropna()  # eliminar filas con datos nulos



# 2.2 Estandarizar formato de fecha (24-10-2021)--> (2022-05-08)
merged_df['invoice_date'] = pd.to_datetime(merged_df['invoice_date'], errors='coerce')
print(merged_df['invoice_date'])

# 2.3 Calcular monto total por venta
merged_df['total_price'] = merged_df['quantity'] * merged_df['price']

# 2.4 Determinar el modo de pago más frecuente (global)
most_common_payment = merged_df['payment_method'].mode()[0]
print(f"\nModo de pago más frecuente: {most_common_payment}")

# 2.5 Modo de pago más frecuente por género
payment_by_gender = merged_df.groupby('gender')['payment_method'].agg(lambda x: x.mode()[0])
print("\nModo de pago más frecuente por género:")
print(payment_by_gender)

# 2.6 Métodos de pago en el rango etario 25–35 años
payment_age_range = merged_df[(merged_df['age'] >= 25) & (merged_df['age'] <= 35)]
payment_by_age = payment_age_range['payment_method'].value_counts()
print("\nMétodos de pago entre 25 y 35 años:")
print(payment_by_age)

# 2.7 Precios promedio por categoría de productos
avg_price_category = merged_df.groupby('category')['price'].mean().sort_values(ascending=False)
print("\nPrecio promedio por categoría:")
print(avg_price_category)

```

```python
# 3.1 Crear un DataFrame limpio
clean_df = merged_df.copy()

# 3.2 Verificar integridad y duplicados
clean_df = clean_df.drop_duplicates()
print("\nVerificación de duplicados eliminados:", clean_df.duplicated().sum())

# 3.3 Mostrar el nuevo DataFrame limpio
print("\nDataFrame limpio y listo para análisis:")
print(clean_df.head(50))
```
